```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
alamat = "https://en.wikipedia.org/wiki/List_of_brightest_stars"
html = urlopen(alamat)
data = BeautifulSoup(html, 'html.parser')
```

```python
table = data.findAll("table", {"class":"wikitable"})[0]
rows = table.findAll("tr")

for row in rows:
    for cell in row.findAll(["td", "th"]):
        print(cell.get_text())
```

```python
hasil = []
for row in rows:
    info = []
    for cell in row.findAll(["td", "th"]):
        info.append(cell.get_text().strip())
    hasil.append(info)
```

```python
hasil
```

```python
import pandas as pd
d=pd.DataFrame(hasil)
d=d.replace(to_replace='\[.\]', value='',regex=True)
header = d.iloc[0]
df  = pd.DataFrame(d.values[1:], columns=header)

# function supaya output hasil tabel bisa ditampilkan semua dengan scroll
def display_():     
    pd.set_option("display.max_rows", None)     
    from IPython.core.display import display     
    display(df)
display_()
```
